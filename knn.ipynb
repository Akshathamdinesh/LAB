```python
# Import required packages
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn import datasets 
from sklearn.metrics import classification_report, confusion_matrix

# Load dataset
iris = datasets.load_iris() 
print("Iris Dataset loaded...") 

# Split the data into train and test samples
x_train, x_test, y_train, y_test = train_test_split(
    iris.data, iris.target, test_size=0.5, random_state=42)

print("Dataset split into training and testing...")
print("Training data shape:", x_train.shape, "Training labels shape:", y_train.shape)
print("Testing data shape:", x_test.shape, "Testing labels shape:", y_test.shape)

# Print label numbers and names
for i in range(len(iris.target_names)): 
    print("Label", i, "-", iris.target_names[i]) 

# Create KNN classifier object
classifier = KNeighborsClassifier(n_neighbors=3)

# Train the classifier
classifier.fit(x_train, y_train)

# Predict test data
y_pred = classifier.predict(x_test)

# Display predictions
print("\nResults of KNN Classification (K=3):")
for i in range(len(x_test)):
    print("Sample:", x_test[i], 
          "Actual Label:", iris.target_names[y_test[i]], 
          "Predicted Label:", iris.target_names[y_pred[i]])

# Accuracy score
print("\nClassification Accuracy:", classifier.score(x_test, y_test))

# Classification metrics
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

# Scatter plot visualization
x_index = 0  # sepal length
y_index = 1  # sepal width

# Format color bar with correct target names
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])

plt.figure(figsize=(8, 6))
plt.scatter(iris.data[:, x_index], iris.data[:, y_index],
            c=iris.target, cmap=plt.cm.get_cmap('RdYlBu', 3), edgecolor='k')

plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.clim(-0.5, 2.5)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])
plt.title("Iris Dataset - Feature Scatter Plot (Sepal Length vs Sepal Width)")
plt.grid(True)
plt.show()

```
